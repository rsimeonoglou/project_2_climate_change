### Scraping all the tables from web-page to get population data from 1950-2050

"https://en.wikipedia.org/wiki/List_of_countries_by_past_and_projected_future_population"


```python
url = "https://en.wikipedia.org/wiki/List_of_countries_by_past_and_projected_future_population"
population_table_list = pd.read_html(url)
population_table_list
```

```python
# Retrieving table-1 within the tables-list scraped from the web-page

population_1950_1980_df = population_table_list[0]
population_1950_1980_df
```

```python
# Table-1 column headers

population_1950_1980_df.columns
```

```python
# # Retrieving table-2 within the tables-list scraped from the web-page

population_1980_2015_df = population_table_list[1]
population_1980_2015_df
```

```python
# Table-2 column headers

population_1980_2015_df.columns
```

```python
# Retrieving table-3 within the tables-list scraped from the web-page

population_2015_2050_df = population_table_list[2]
population_2015_2050_df
```

```python
# Table-3 column headers

population_2015_2050_df.columns
```

### Cleaning & Transformation

```python
# Merging all three population dataframes

population_1950_2015_df = population_1950_1980_df.merge(population_1980_2015_df, how='left', on='Country (or dependent territory)')
population_1950_2050_df = population_1950_2015_df.merge(population_2015_2050_df, how='left', on='Country (or dependent territory)')
population_1950_2050_df.columns
```

```python
# Renaming columns

population_1950_2050_df.rename(columns={"Country (or dependent territory)": "Country", "%_x": "Growth 1950-1955",
                                                                                    "%.1_x": "Growth 1955-1960", 
                                                                                   "%.2_x": "Growth 1960-1965",
                                                                                   "%.3_x": "Growth 1965-1970",
                                                                                    "%.4_x": "Growth 1970-1975",
                                                                                    "%.5_x": "Growth 1975-1980",
                                                                                    "%_y": "Growth 1980-1985",
                                                                                    "%.1_y": "Growth 1985-1990",
                                                                                     "%.2_y": "Growth 1990-1995",
                                                                                   "%.3_y": "Growth 1995-2000",
                                                                                    "%.4_y": "Growth 2000-2005",
                                                                                   "%.5_y": "Growth 2005-2010",
                                                                                   "%.6_x": "Growth 2010-2015",
                                                                                    "%": "Growth 2015-2020",
                                                                                    "%.1": "Growth 2020-2025",
                                                                                   "%.2": "Growth 2025-2030",
                                                                                   "%.3": "Growth 2030-2035",
                                                                                    "%.4": "Growth 2035-2040",
                                                                                   "%.5": "Growth 2040-2045",
                                                                                   "%.6_y": "Growth 2045-2050"}, 
                                                                                    inplace =True)

population_1950_2050_df
```

```python
# Retrieving lattitude & longitude using googlemaps API

base_url = "https://maps.googleapis.com/maps/api/geocode/json"

country = population_1950_2050_df["Country"].tolist()
lat=[]
lng=[]
for i in country:
    target_url = ('https://maps.googleapis.com/maps/api/geocode/json?address={0}&key={1}').format(i, gkey)
    response = requests.get(target_url).json()
    lattitude = response["results"][0]["geometry"]["location"]["lat"]
    longitude = response["results"][0]["geometry"]["location"]["lng"]
    lat.append(lattitude)
    lng.append(longitude)
```

```python
# Populating 2 new-columns Lat & Lng into the same population dataframe using col-index

population_1950_2050_df.insert(1, 'Lat', lat)
population_1950_2050_df.insert(2, 'Lng', lng)
```

```python
# population_1950_2050_df.drop(['1950'],axis=1, inplace=True)

population_1950_2050_df
```

```python
# Changing all year columns-cells to into a single column 'Year' & values to 'Population'
df = population_1950_2050_df
df1 = df.melt(id_vars=["Country", "Lat", "Lng"], 
        value_vars=['1955','1960','1965','1970','1975','1980','1985','1990','1995','2000','2005','2010','2015','2020','2025','2030','2035','2040','2045','2050'],
        var_name='Year',
        value_name='Population')
                    
df1
# df3=df1.groupby(['Country','Year']).sum()
# # df3 = df1.groupby("Year").mean()
# # # df2.drop(['World'],axis=0, inplace=True)
# df3
```

```python
# Changing all growth columns-cells to into a single column 'growth' & their values as 'Rate'

df2 = df.melt(id_vars=["Country"], 
        value_vars=['Growth 1950-1955','Growth 1955-1960',
        'Growth 1960-1965','Growth 1965-1970', 'Growth 1970-1975', 'Growth 1975-1980', 
        'Growth 1980-1985', 'Growth 1985-1990',
       'Growth 1990-1995', 'Growth 1995-2000',
       'Growth 2000-2005', 'Growth 2005-2010', 
       'Growth 2010-2015', 'Growth 2015-2020', 
       'Growth 2020-2025', 'Growth 2025-2030', 
       'Growth 2030-2035', 'Growth 2035-2040', 
       'Growth 2040-2045', 'Growth 2045-2050'],
        var_name='Period',
        value_name='Rate-Change')

# Merging 'Period' column

df2.drop(['Period'],axis=1, inplace=True)
df2

df1['Rate'] = df2['Rate-Change']
df1
```
