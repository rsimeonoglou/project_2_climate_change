```python
import pandas as pd
import datetime as dt
from datetime import timedelta
from sqlalchemy import create_engine

pd.set_option("display.max_rows", 50, "display.max_columns", None)
```

```python
global_df = pd.read_csv("GlobalTemperatures.csv")
global_df
```

```python
global_df['dt'] = pd.to_datetime(global_df['dt'])
global_df['Year'] = global_df["dt"].dt.year
global_df['Month'] = global_df["dt"].dt.month
global_df['Day'] = global_df["dt"].dt.day

#datetime changed 1900-1970 to 2000s, need to correct
df_before_1900 = global_df.iloc[:1800]
df_after_1970 = global_df.iloc[2652:]

df_1900_to_1970 = global_df.iloc[1800:2652]         
df_1900_to_1970["Year"] = df_1900_to_1970["Year"]-100
df_1900_to_1970['dt'] = df_1900_to_1970["Year"].astype(str) + "-" + df_1900_to_1970["Month"].astype(str) + "-" + df_1900_to_1970["Day"].astype(str)
df_1900_to_1970['dt'] = pd.to_datetime(df_1900_to_1970['dt'])

global_df = pd.concat([df_before_1900,df_1900_to_1970,df_after_1970],ignore_index=True)
global_df
```

```python
global_df = global_df.loc[(global_df["Year"] > 1792)]
global_df = global_df.loc[(global_df["Year"] < 2014)]
global_df
```

```python
#celcius to F calc (0Â°C Ã— 9/5) + 32 

global_df["land_avg_fahrenheit"] = (global_df["LandAverageTemperature"] * 9/5) +32
global_df.reset_index(inplace=True) 
global_df
```

```python
global_df_groupby = global_df.groupby("Year").mean()
global_df_groupby = global_df_groupby[["LandAverageTemperature","land_avg_fahrenheit"]].reset_index()
global_df_groupby

global_df_groupby.to_csv("global_group.csv")
```

```python
global_df_groupby["Period"] = 0
counter = 0

for index, row in global_df_groupby.iterrows():
    counter = counter + 1
    if counter == 11:
        counter = 1
        global_df_groupby.loc[index, 'Period'] = counter
    else:
        global_df_groupby.loc[index, 'Period'] = counter

global_df_groupby(221)
```

```python
global_df_decades = global_df_groupby.loc[(global_df_groupby["Period"]==1) |
                                         (global_df_groupby["Period"]==10)]
global_df_decades

global_df_decades["temp_change"] = 0
global_df_decades["year_range"] = 0
temp_change = 0
beg_temp = 0

for index, row in global_df_decades.iterrows():
    
    if row["Period"] == 1:
        beg_temp = row["land_avg_fahrenheit"]
        beg_year = int(row["Year"])
        global_df_decades.loc[index, 'temp_change'] = 0
    else:
        temp_change = row["land_avg_fahrenheit"] - beg_temp
        global_df_decades.loc[index, 'temp_change'] = temp_change
        end_year = int(row["Year"]+1)
        year_range = "{} - {}".format(beg_year,end_year)
        global_df_decades.loc[index, 'year_range'] = year_range

global_df_decades
```

```python
global_df_decades = global_df_decades.loc[global_df_decades['Period']==10]
global_df_decades

global_df_decades.to_csv("decades.csv")
```
