```python
#namibia lat and long were null in location dataset, so manually input
namibia = country_df.loc[country_df["Country"]=="Namibia"]
namibia["latitude"] = -22.967062
namibia["longitude"] = 18.4929993

country_df_merged = pd.merge(country_df,country_locations,on="Country")
country_df_merged = pd.concat([country_df_merged,namibia],ignore_index=True)
country_df_merged

```

```python
#Remove countries with data that does not go back to 1900
country_df_merged = country_df_merged.loc[(country_df_merged["Country"]!= "Antarctica") &
                                          (country_df_merged["Country"]!= "French Southern And Antarctic Lands") &
                                         (country_df_merged["Country"]!= "Heard Island And Mcdonald Islands") &
                                         (country_df_merged["Country"]!= "Northern Mariana Islands") &
                                         (country_df_merged["Country"]!= "Guam")]

country_df_merged_year = country_df_merged.groupby(["Country","Year"]).mean().reset_index()
country_df_merged_year = country_df_merged_year.loc[(country_df_merged_year["Year"] > 1899) &
                                                   (country_df_merged_year["Year"] < 2013)].reset_index(drop=True)
country_df_merged_year
```

```python
beg_temp = 0
country_df_merged_year["year change"] = 0
country_df_merged_year["running_total_change"] = 0

for index, row in country_df_merged_year.iterrows():
    if row["Year"] == 1900:
        country_df_merged_year.loc[index, 'year change'] = 0
        country_df_merged_year.loc[index, 'running_total_change'] = 0
        beg_temp = row["avg_temp_F"]
        running_total = 0
    else:
        change = row["avg_temp_F"] - beg_temp
        running_total = running_total + change
        country_df_merged_year.loc[index, 'year change'] = change
        country_df_merged_year.loc[index, 'running_total_change'] = running_total
        beg_temp = row["avg_temp_F"]
        
country_df_merged_year.head(20)
```
